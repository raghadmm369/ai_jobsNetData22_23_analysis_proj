# About Dataset
Download date: (7-1-2024)  
These data were obtained from [ai-jobs.net]( https://ai-jobs.net/salaries/download/) , a website specialized in artificial intelligence, machine learning, and data science jobs.

The data contains 11 columns as follows:
- **work_year:** The year the salary was paid.
- **experience_level:** <br> The experience level in the job during the year with the following possible values:
  - **EN:** Entry-level / Junior. 
  - **MI:** Mid-level / Intermediate.   
  - **SE:** Senior-level / Expert.  
  - **EX:** Executive-level / Director.
- **employment_type:** <br> The type of employement for the role: 
  - **PT:** Part-time. 
  - **FT:** Full-time. 
  - **CT:** Contract. 
  - **FL:** Freelance.
- **job_title:** The role worked in during the year.
- **salary:** The total gross salary amount paid.
- **salary_currency:** The currency of the salary paid as an ISO 4217 currency code.
- **salary_in_usd:** The salary in USD (FX rate divided by avg. USD rate of respective year) via statistical data from the BIS and central banks.
- **employee_residence** Employee's primary country of residence in during the work year as an ISO 3166 country code.
- **remote_ratio:** <br> The overall amount of work done remotely, possible values are as follows: 
  - **0:** No remote work (less than 20%).
  - **50:** Partially remote/hybird. 
  - **100:** Fully remote (more than 80%). 
- **company_location** The country of the employer's main office or contracting branch as an ISO 3166 country code.
- **company_size:** <br> The average number of people that worked for the company during the year:
  - **S:** less than 50 employees (small).
  - **M:** 50 to 250 employees (medium). 
  - **L:** more than 250 employees (large).

## ما هي نسبة الاختلاف في تكرار الوظائف بين عامي 2022 و 2023؟
###### What is the percentage difference in the recurrence of jobs between 2022 and 2023?

```python
jobs_in23_in22= jobs_23[jobs_23.isin(jobs_22)]
jobs_in23_in22= df_2022_2023[df_2022_2023.job_title.isin(jobs_in23_in22)]
pivot1= jobs_in23_in22.pivot_table(index=['job_title'], columns=['work_year'],values='company_location',aggfunc='count')
count_22= pivot1[2022].values
count_23= pivot1[2023].values
# حساب التغير بالنسبة المئوية
#Calculate the percentage change
pivot1['percentage_change']= (count_23-count_22)/count_22*100
pivot1= pivot1.sort_values(by='percentage_change',ascending=False)
pivot1.style.background_gradient(axis=0, cmap='viridis')
```

#####  نلاحظ زيادة عدد الوظائف التي تمت في عام 2023 مقارنةً بعام 2022 بنسبة لا تقل عن 25٪، باستثناء 6 وظائف حيث بقي عددها ثابتًا، و 4 وظائف فقط انخفضت أعدادها في عام 2023 وهي كالتالي
 We notice an increase in the number of jobs filled in 2023 compared to 2022 by at least 25%, with the exception of 6 jobs where their count remained the same. However, there are only 4 jobs that decreased in count in 2023, as follows:"

## ماهي نسبة تغير متوسط رواتب الوظائف بين عام 2022 و2023؟
###### What is the percentage change in the median salaries of jobs between 2022 and 2023?

```python
jobs_in23_in22=jobs_23[jobs_23.isin(jobs_22)]
jobs_in23_in22=df_2022_2023[df_2022_2023.job_title.isin(jobs_in23_in22)]
pivot2=jobs_in23_in22.pivot_table(index=['job_title'], columns=['work_year'], values='salary_in_usd',aggfunc=np.median).round(2)
salary_22=pivot2[2022].values
salary_23=pivot2[2023].values
# حساب التغير بالنسبة المئوية
#Calculate the percentage change
pivot2['percentage_change']=(salary_23-salary_22)/salary_22*100
pivot2=pivot2.sort_values(by='percentage_change',ascending=False)
pivot2.style.background_gradient(axis=0, cmap='viridis')
```
